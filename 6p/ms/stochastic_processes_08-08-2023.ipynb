
## Rolando dados
---



```python
import numpy as np

num_lados = 6
num_rolagens = 100

# Simula rolagens
vetor_rolagens = np.random.randint(1, num_lados + 1, size=num_rolagens)
# Calcula media
print(vetor_rolagens)
media = np.mean(vetor_rolagens)

print("Media de 100 rolagens de dados:", media)
```

## Lan√ßamentos de moedas

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize starting position and list of positions
position = 0
positions = [0]

# Generate 50 random steps of size 1 and update position
for i in range(200):
    step = np.random.choice([-1, 1])
    position += step
    positions.append(position)

# Plot the random walk
plt.plot(positions)
plt.title("Random Walk (50 Steps of Size 1)")
plt.xlabel("Step Number")
plt.ylabel("Position")
plt.show()
```

## Lan√ßamento de dados com soma

```python
import numpy as np
import matplotlib.pyplot as plt

results = []

# Preenche vetor com rolagens de dados
for i in range(1000):
    results.append(np.random.randint(1, 7) + np.random.randint(1, 7) + np.random.randint(1, 7))

print(results)

# Plotar histograma
plt.hist(results, bins=range(1, 20), align='left', rwidth=0.8, edgecolor='black')
plt.xlabel('Result of 2d6 Roll')
plt.ylabel('Frequency')
plt.title('Histogram of 2d6 Rolls')
plt.xticks(range(1, 20))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

results = []

for i in range(100000):
    
    if np.random.rand() < 0.5:
        
        if np.random.rand() < 2/9:
            # peguei branca heads
            results.append('BH')
        else:
            # peguei preta heads
            results.append('PH')
    else:
        if np.random.rand() < 5/11:
            # peguei branca tails
            results.append('BT')
        else:
            # peguei preta tails
            results.append('PT')
      
a, counts = np.unique(results, return_counts=True)
print(a, counts)

plt.bar(a, counts)

print(counts[0]/(counts[0] + counts[1]))
```
